## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Parameters

```python
N = 10
domain = [1,5]

f = lambda t,x: np.sin(t)
analytic = lambda t: -np.cos(t) + 2 + np.cos(1)
```

## Algorithm


```python
def runge_kutta_stepping(t: np.array, f: callable, x_0: float) -> np.array:
    '''
    _summary_

    Args:
        t: _description_
        f: _description_
        x_0: _description_

    Returns:
        _description_
    '''

    dt = t[1] - t[0]
    x = np.zeros_like(t)

    # initial condition
    x[0] = x_0

    for n in range(0, len(t)-1):

        # single runge kutta step
        k_1 = f(t[n], x[n])
        k_2 = f(t[n]+dt/2, x[n]+(dt*k_1)/2)
        k_3 = f(t[n]+dt/2, x[n]+(dt*k_2)/2)
        k_4 = f(t[n]+dt, x[n]+dt*k_3)
        
        x[n+1] = x[n] + (dt/6)*(k_1 + 2*k_2 + 2*k_3 + k_4)

    return x
```

```python
t = np.linspace(*domain, num=N)

x_sol = runge_kutta_stepping(t, f=f, x_0=2)
```

```python
plt.plot(t, x_sol, "o--")
plt.plot(np.linspace(1,5, num=500), analytic(np.linspace(1,5, num=500)))

plt.grid()
plt.show()
```
