## Algorithm


```python
t = np.linspace(1,5, num=12)
dt = t[1] - t[0]

x = np.zeros_like(t)

x[0] = 2

f = lambda t,x: -2*t*x

for n in range(0, len(x)-1):
    k_1 = f(t[n], x[n])
    k_2 = f(t[n]+dt/2, x[n]+(dt*k_1)/2)
    k_3 = f(t[n]+dt/2, x[n]+(dt*k_2)/2)
    k_4 = f(t[n]+dt, x[n]+dt*k_3)
    
    x[n+1] = x[n] + (dt/6)*(k_1 + 2*k_2 + 2*k_3 + k_4)
```

```python
plt.plot(t, x, "o-")
plt.plot(np.linspace(1,5, num=500), 2*np.exp(1-np.linspace(1,5, num=500)**2))

plt.show()
```
