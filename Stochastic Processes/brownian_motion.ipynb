## Parameters

```python
number_steps = 1000
standard_deviation = 1
```

## Algorithm


```python
def BM(x_0: float, SD: float, steps: int) -> tuple[float, list, int]:
    '''

    '''
    

    states=np.zeros(steps)
    states[0] = x_0

    for i in range(steps-1):
        
        move = np.random.normal(0, SD)
        states[i+1] = states[i] + move

    return np.array(range(steps)), states
```

```python
def BM_distribution(x_0: float, SD: float, steps: int, samples: int) -> tuple[float, list, int]:
    '''
    Compute the distribution of the Wiener process
    '''

    mu = 0
    sigma = SD
    
    r = 1/steps
    t = steps*r

    distribution_samples = np.zeros(samples)

    for i in range(samples):
        
        wiener = np.random.normal(0, t)
        distribution_samples[i] = x_0 + mu*(t/r) + (np.sqrt(sigma)/np.sqrt(r))*wiener

    return distribution_samples
```

## Results Visualization & Behaver Analysis


```python

fig, ax = plt.subplots(figsize=(17,7))


for i in range(1000):
    steps, states = BM(x_0=0, SD=standard_deviation, steps=number_steps)
    ax.plot(steps, states)

rw_dis = BM_distribution(x_0=0, SD=standard_deviation, steps=number_steps, samples=int(1e4))
ax.hist(rw_dis, bins=50, orientation="horizontal", bottom=number_steps, edgecolor='black', facecolor='lightblue', alpha=0.4, label="Random Walk Distribution")


# Set labels on joint
ax.set(xlabel='# Moves', ylabel='Position')
ax.grid(True)
ax.legend()

# Set labels on marginals
plt.show()
```
